```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# membaca data dengan format csv
data = pd.read_csv('Pendataan_Peserta_Magang_CEO_HMSD_2024_2.csv')
data = data.rename(columns={"3. Rating Kesulitan Magang (1-5, 5 semakin sulit)": "Rating Kesulitan"})
print(data)
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data["Rating Kesulitan"], bins=5, kde=True, color="#006d75")
plt.title("Distribusi Rating Kesulitan Magang")
plt.xlabel("Rating Kesulitan")
plt.ylabel("Frekuensi")
plt.show()
```

```python
colors = sns.color_palette("muted")
plt.figure(figsize=(8, 8))
data["Rating Kesulitan"].value_counts().plot.pie(autopct='%1.1f%%', colors=colors)
plt.title("Proporsi Rating Kesulitan Magang")
plt.ylabel("")  # Menghilangkan label sumbu y
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="Rating Kesulitan", data=data, palette=colors)
plt.title("Frekuensi Setiap Rating Kesulitan Magang")
plt.xlabel("Rating Kesulitan")
plt.ylabel("Frekuensi")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data["Rating Kesulitan"], fill=True, color="#3FBAD8", alpha=0.6)
plt.title("Density Plot untuk Rating Kesulitan Magang")
plt.xlabel("Rating Kesulitan")
plt.ylabel("Kepadatan")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=data["Rating Kesulitan"], color="#ffa500")
plt.title("Boxplot Rating Kesulitan Magang")
plt.xlabel("Rating Kesulitan")
plt.show()
```
