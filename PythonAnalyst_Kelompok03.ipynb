```python
# cut -> melakukan pemotongan berdasarkan kelas tsb
# income ->
# 1. gaji <= 40
# 2. gaji di antara 40 - 70
# 3. gaji > 70
data_buyer['Income_grouped'] = pd.cut(data_buyer['Annual Income (k$)'], bins=[0, 40, 70, 200])
data_buyer
```

```python
data_buyer['spending_score_grp'] = pd.cut(data_buyer['Spending Score (1-100)'], bins=[0, 40, 60, 100])
data_buyer
```

```python
data_buyer.groupby(['spending_score_grp', 'Income_grouped'])['CustomerID'].count()
```

```python
pd.pivot_table(data=data_buyer, index='spending_score_grp', columns='Income_grouped', 
               values='Annual Income (k$)', aggfunc='mean')
```

```python
# 5 group 
# rules 
# 1. Low Income, Low Score -> income <= 40, spending <= 40
# 2. High Income, Low Score -> Income >= 70, spending <= 40
# 3. Low Income, High Score -> Income <= 40, spending >= 60
# 4. High Income, High Score -> Income >= 70, spending > 60
# 5. Med Income, Med Score -> income 40 - 70, spending 40 - 60

# fungsi 
def jenis_grup(income, spending_score):
    if income <= 40 and spending_score <= 40:
        jenis = 'Low Income, Low Score'
    elif income <= 40 and spending_score > 60: 
        jenis = 'Low Income, High Score'
    elif income > 70 and spending_score >= 60:
        jenis = 'High Income, High Score'
    elif income > 70 and spending_score <= 40:
        jenis = 'High Income, Low Score'
    elif income >= 40 and income <= 70 and spending_score >= 40 and spending_score <= 60:
        jenis = 'Med Income, Med Score'
    else:
        jenis = 'Unknown'

    return jenis
```

```python
# .apply(lambda x: fungsi_yg_mau_diapply)
data_buyer.apply(lambda x: jenis_grup(x['Annual Income (k$)'], x['Spending Score (1-100)']), axis=1)
```

```python
data_buyer['segment_user'] = data_buyer.apply(lambda x: jenis_grup(x['Annual Income (k$)'],x['Spending Score (1-100)']), axis=1)
data_buyer
```
